```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = ['model', 'episode', 'Null', 'score', 'avg. score', 'max. avg. score', 'Null', 'TAU',  'LR_ACTOR', 'LR_CRITIC', 'fc_1', 'fc_2']

```

```python
exp1 = pd.read_csv('results.txt', header = None, names = cols )
```

```python
fig = plt.figure(figsize=(8, 7))
for i in range(1, 13):
    x = exp1[exp1.model == i]['episode']   
    y = exp1[exp1.model == i]['avg. score']
    plt.plot(x, y, label = 'model no. '+ str(i))
    plt.legend(loc='upper right')
plt.title("Experiment 1: All models until solved")
plt.xlabel('Episode number')
plt.ylabel('Avg score')
plt.show()
```
