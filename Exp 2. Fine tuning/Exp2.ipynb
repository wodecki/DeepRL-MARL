```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = ['model', 'episode', 'Null', 'score', 'avg. score', 'max. avg. score', 'Null', 'TAU',  'LR_ACTOR', 'LR_CRITIC', 'fc_1', 'fc_2']

```

```python
exp2 = pd.read_csv('results.txt', header = None, names = cols )
```

```python
fig = plt.figure(figsize=(8, 7))
for i in range(1, 5):
    x = exp2[exp2.model == i]['episode']   
    y = exp2[exp2.model == i]['avg. score']
    plt.plot(x, y, label = 'model no. '+ str(i))
    plt.legend(loc='upper right')
plt.title("Experiment 1: All models until solved")
plt.xlabel('Episode number')
plt.ylabel('Avg score')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 7))
x = exp2[exp2.model == 4]['episode']   
y = exp2[exp2.model == 4]['avg. score']
plt.plot(x, y, label = 'The best model ')
plt.legend(loc='upper right')
plt.title("The best model")
plt.xlabel('Episode number')
plt.ylabel('Avg score')
plt.show()
```
